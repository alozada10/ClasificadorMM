```python
import numpy as np
import matplotlib.pyplot as plt

#Se cargan los datos, se define la función promedio y la función desviación estándar

def prom(lista):
    a=0.0
    for i in range(len(lista)):
        a += lista[i]/len(lista)
    return a


def desv(lista):
    resp= 0.0
    pm = prom(lista)
    for i in range(len(lista)):
        dis = lista[i] - pm
        resp += dis**2
    return (resp/len(lista))**0.5



try1 = np.loadtxt('Try1.csv', delimiter = ',')
try2 = np.loadtxt('Try2.csv', delimiter = ',')
try3 = np.loadtxt('Try3.csv', delimiter = ',')

try4 = np.loadtxt('Try4.csv', delimiter = ',')
try5 = np.loadtxt('Try5.csv', delimiter = ',')
try6 = np.loadtxt('Try6.csv', delimiter = ',')

try7 = np.loadtxt('Try7.csv', delimiter = ',')
try8 = np.loadtxt('Try8.csv', delimiter = ',')
try9 = np.loadtxt('Try9.csv', delimiter = ',')

#try10 = np.loadtxt('Try10.csv', delimiter = ',')
#try11 = np.loadtxt('Try11.csv', delimiter = ',')
#try12 = np.loadtxt('Try12.csv', delimiter = ',')
```

```python
#Filtro para cada imágen. CCC es el archivo con la circularidad, ttt el archivo con todos los datos de la imagen
#perim el archivo con el perímetro, area con el área, h con el Height, y w con el width.
#Retorna la lista con el número de los candidatos. Es decir, si aparece '2340' hay que buscar en el archivo con 
#los datos de la imagen la partícula número '2340'.


def fs(CCC,tt,perimN,areaN,hN,wN):
    
    f1 = []
    for i in range(len(CCC)):
        if((CCC[i]<= pCIR+2*dCIR) & (CCC[i]>= pCIR-2*dCIR)):
            f1.append(CCC[i])
        
    resp = []
    cent=1
    for i in range(len(tt)):
        for j in range(len(f1)):
            if((tt[i][16]==f1[j])& (i != cent)):
                cent = i
                resp.append(i)
            
    f2 = []
    for i in range(len(resp)):
        a = resp[i]
        if((perimN[a]>= pperim-2*dperim)&(perimN[a]<= pperim+2*dperim)):  
            f2.append(a)
            
    f3 = []
    for i in range(len(f2)):
        a = f2[i]
        if((hN[a]>= pH -2*dH)&(hN[a]<= pH +2*dH)):
            f3.append(a)
        
            
    f4 = []
    for i in range(len(f3)):
        a = f3[i]
        if((wN[a]>= pW -2*dW)&(wN[a]<= pW +2*dW)):
            f4.append(a)
            
    f5 = []
    for i in range(len(f4)):
        a = f4[i]
        if((areaN[a] >= parea-2*darea)&(areaN[a] <= parea+2*darea)):  
            f5.append(a)
               
            
    return np.array(f5)
```
